## Poroƒçilo vaja 4.3
#### Analiza variance: vpliv treninga na VO2max

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pingouin as pg
from statsmodels.stats.anova import AnovaRM

path = os.getcwd()
df = pd.read_spss(os.path.join(path, 'data\\vo2max-anova-repeated-2grps-gen.sav'))
df.skupina = pd.Categorical(df.skupina.map({1: '1', 2: '2'}))
df = df.rename(columns={'VO2max1': 'week 00',
                   'VO2max2': 'week 05',
                   'VO2max3': 'week 10',
                   'VO2max4': 'week 15',
                   'VO2max5': 'week 20',
                   'skupina': 'group'
                  })
df.head()
time = ['week 00', 'week 05', 'week 10', 'week 15', 'week 20']
```

```python
df_pivot = df.pivot_table(
    values= time,
    index='group',
    aggfunc={'count','mean', 'std', 'var'},
    observed=True
)
styled_columns = df_pivot.columns.to_list()
styles_h = [
    {'selector': f'th.col{i}', 'props': [('border-right', '1px solid gray')]}
    for i in range(len(styled_columns))
]
n_cols = len(df_pivot.columns)

styles_c = [
    {'selector': f'td.col{i}', 'props': [('border-right', '1px solid gray')]}
    for i in range(n_cols)
]

display(df_pivot.style
        .set_caption('Table 1: Descriptive statistics')
        .set_table_styles([
            {'selector': 'caption', 'props': [
                ('font-size', '16px'),
                ('font-weight', 'bold'),
                ('border-bottom', '1px solid black'),
                ('width', '100%')
            ]},
            {'selector': 'td', 'props': [('text-align', 'left')]}
        ] + styles_h + styles_c)
        .format(precision=2)
)
```

```python
grouped_median = df.groupby('group', observed=True)[time].mean()

grouped_median = grouped_median.T

grouped_median.plot(kind='line', marker='o')
plt.title('Mean VO2max by group')
plt.xlabel('Time [weeks]')
plt.ylabel('VO2max [mL/kg*min]')
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```

Iz opisne statistike opazimo, da obstajajo razlike med vrednostmi VO2max med skupinama, prav tako pa obstajajo razlike med vrednostmi VO2max znotraj posamezne skupine. Dokaz, o tem ali je razlika med VO2max znotraj skupine statistiƒçno znaƒçilna bomo dokazali z ANOVO ponavljajoƒçih meritev. 

**HO:** Med meritvami VO2max v razliƒçnih tednih ne obstajajo statistiƒçno znaƒçilne razlike.

**H1:** Med meritvami V=2max v razliƒçnih tednih obstajajo statistiƒçno znaƒçilne razlike.

```python
df['subject'] = df.index

df_long = pd.melt(df,
                  id_vars=['subject', 'group'],
                  value_vars=time,
                  var_name='week',
                  value_name='vo2max')


df_long['week'] = df_long['week'].str.extract(r'(\d)').astype(int)
g_1 = df_long[df_long['group'] == '1'].drop(columns='group')
g_2 = df_long[df_long['group'] == '2'].drop(columns='group')

aov1 = pg.rm_anova(data=g_1, dv='vo2max', within='week', subject='subject', correction=True, detailed=True)
aov2 = pg.rm_anova(data=g_2, dv='vo2max', within='week', subject='subject', correction=True, detailed=True)
anova_rm = pg.rm_anova(data=df_long, dv='vo2max', within='week', subject='subject', correction=True, detailed=True)

display(aov1.style.set_caption('Repeated measures ANOVA: group 1').set_table_styles([
            {'selector': 'caption', 'props': [
                ('font-size', '16px'),
                ('font-weight', 'bold'),
                ('border-bottom', '1px solid black'),
                ('width', '100%')
            ]}]))
display(aov2.style.set_caption('Repeated measures ANOVA: group 2').set_table_styles([
            {'selector': 'caption', 'props': [
                ('font-size', '16px'),
                ('font-weight', 'bold'),
                ('border-bottom', '1px solid black'),
                ('width', '100%')
            ]}]))
display(anova_rm.style.set_caption('Repeated measures ANOVA: between groups').set_table_styles([
            {'selector': 'caption', 'props': [
                ('font-size', '16px'),
                ('font-weight', 'bold'),
                ('border-bottom', '1px solid black'),
                ('width', '100%')
            ]}]))
```

**Interpretacija:** Iz rezultatov v tabeli lahko vidimo, da je sferiƒçnost v primeru obeh skupin kr≈°ena, zato smo popravili p vrednost ANOVE ponovljivih meritev z Greenhouse-Geisser popravkom. Opazimo, da ƒças treninga statistiƒçno znaƒçilno vpliva na vrednosti VO2max pri obeh skupinah, kot tudi, da je statistiƒçno znaƒçilna razlika med vrednostmi VO2max med skupinama. Iz zgornjega grafa lahko razberemo, da je napredek pri skupini 1 veƒçji kot pri skupini 2. Opravimo tudi post-hoc analizo, da ugotovimo v katerem tednu se zaƒçne kazati statistiƒçno znaƒçilna razlika v vrednostih VO2max za prvo skupino in za drugo skupino.

```python
posthoc1 = pg.pairwise_tests(data=g_1, dv='vo2max', within='week', subject='subject', parametric=True, padjust='none')
display(posthoc1)
posthoc2 = pg.pairwise_tests(data=g_2, dv='vo2max', within='week', subject='subject', parametric=True, padjust='none')
display(posthoc2)
```
