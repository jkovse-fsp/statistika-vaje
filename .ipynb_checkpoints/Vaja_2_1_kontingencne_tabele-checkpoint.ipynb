#### Kontingenčne tabele: Preučevanje faktorjev tveganja za poškodbe pri rekreativnem teku

Preučujemo faktorje tveganja za poškodbe pri rekreativnem teku. Podatke smo zbrali v datoteki: *running_injury.sav*

1. Ugotovite, katere faktorje, ki smo jih zajeli v naših podatkih, lahko povezujemo s poškodbami pri teku. Analize izvedite:
   - s kontingenčnimi tabelami,
   - z ustreznimi grafi,
   - z ustreznimi statističnimi testi,
   - s statistikami tveganja.
- Ustrezno predstavite rezultate v eni skupni tabeli skupaj z vsemi statistikami.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
from scipy.stats import shapiro, chi2_contingency
from scipy.stats.contingency import odds_ratio
import pingouin as pg

path = os.getcwd()
df = pd.read_spss(path + "\\data\\running_injury.sav")
df.rename(columns={'BMI26': 'BMI<26', 'poskodba': 'trajanje_poskodbe', 'poskodba_da_ne': 'poskodba', 'izkusnje10': 'izkusnje<10'}, inplace=True)
df.head()
```

#### Pregled normalnosti porazdelitve spremenljivk 'starost', 'tedenski_ki', 'povp_hitrost' z uporabo Shapiro-Wilk testa.
Ta test se uporablja če imamo meritev n < 50. Če imamo meritev n > 250 bi se naj uporabljal Kolmogorov-Smirnov test. Smernice so, naj vedno uporabimo Shapiro-Wilk test.

Skalarne spremenljivke bomo spremenili v kategorične. Najprej preverimo normalnost porazdelitve, potem se odločimo ali bomo kot mejo razdelitve v posamezno kategorijo vzeli povprečje (normalna porazdelitev) ali mediano (ne-normalna porazdelitev). Skalarna spremenljivke se pretvori v kategorično v smislu **<= x / > x**, kjer je x mediana ali povprečje.

```python
# Check for normality of continuous variables
for var in ['starost', 'tedenski_km', 'povp_hitrost']:
    variable = df[var].dropna()
    stat, p = shapiro(variable)
    print(f'Shapiro-Wilk test statistic for {var}: {stat:.4f}')
    print(f'                              p-value: {p:.4f}')
    if p < 0.05:
        print(f'Variable **{var}** is not normally distributed. Using MEDIAN to halve it.')
        print(f'') 
        median = variable.median()
        df[f'{var}<{median:.1f}'] = pd.Categorical(np.where(df[var] < median, f'< {median:.1f}', f'> {median:.1f}'))
    else:
        print(f'Variable **{var}** is normally distributed. Using MEAN to halve it.')
        print('')
        mean = variable.mean()
        df[f'{var}<{mean:.1f}'] = pd.Categorical(np.where(df[var] < mean, f'< {mean:.1f}', f'> {mean:.1f}'))

df.head()
```

#### Kontingenčna tabela
Kontingenčna tabela je tabela frekvenc, ko križamo dve kategorijski spremenljivki. V celicah tabele podamo število primerkov pri posamezni kombinaciji posameznih kategorij. Spodaj s pomočjo hi-kvadrat testa izračunamo veljavnost ničte hipoteze H0, ki pravi, da med dejavnikoma ni povezave. Magnitudo povezanosti nato izračunamo s Cramerjevim V koeficientom (1: popolna povezanost, 0: popolna nepovezanost). 

Če sta kategorijski spremenljivki povezani, izračunamo razmerje obete za poškodbo. Če je razmerje obetov:
1. = 1; ni razlike v obetih,
2. < 1; manjše tveganje,
3. \> 1; večje tveganje.

```python
# 
results = []
df['poskodba'] = df['poskodba'].astype('category')
for category in ['spol', 'BMI<26', 'stopalo', 'izkusnje<10', 'ogrevanje', 'vaje', 'starost<44.6', 'tedenski_km<21.1', 'povp_hitrost<11.2']:
    table = pd.crosstab(df[category], df['poskodba'])
    out, observed, expected = pg.chi2_independence(data=df, x=category, y='poskodba')
    
    res = expected.iloc[0]
    
    if res['pval'] < 0.05:
        print(f'=====================================================\nHi-kvadrat test za spremenljivko {category} vs poskodba')
        print('Kontingenčna tabela:\n')
        display(table)
        print(f'Hi-kvadrat: {res['chi2']:.2f},     p-vrednost: {res['pval']:.4f},      Prostostne stopnje: {res['dof']}\n')
        
        print(f'====> Statistično značilna povezanost poškodb in spremenljivke {category}!')
        df[category] = df[category].astype('category')
        # calculate odds_ratio
        odds = odds_ratio(table)
        stat = odds.statistic
        low, high  = odds.confidence_interval(confidence_level=.95)

        print(f'\nObeti za poškodbo glede na {category}: {stat:.2f} (CI={low:.2f}, {high:.2f})\n')
        if stat < 1:
            print(f'Možnost poškodbe je manjša, če smo izpostavljeni faktorju {category}')
        else:
            print(f'Možnost poškodbe je večja, če smo izpostavljeni faktorju {category}')
    
        sns.countplot(x=category, hue='poskodba', data=df)
        plt.title(f'Poškodba glede na {category}')
        plt.show()
    
        results.append({
            'Spremenljivka': category,
            'Chi2': res['chi2'],
            'p-value': f"{res['pval']:.3f}",
            'dof': res['dof'],
            'Cramer\'s V': res['cramer'],
            'Odds ratio (CI95%)': f'{stat:.3f} ({low:.3f}, {high:.3f})'
        })
```

**Končni rezultati statistične analize so predstavljeni v tabeli 1**

```python
results_df = pd.DataFrame(results)
results_df = results_df.set_index('Spremenljivka')
results_df = results_df.rename(index={'BMI26': 'BMI < 26', 'tedenski_km21.1': 'km/teden > 21.1'})
results_df.style.set_caption(f'Tabela 1: Rezultati testov kontingenčne tabele')\
          .set_table_styles([
              {'selector': 'caption', 'props': [('font-size', '16px'), 
                                                ('font-weight', 'bold'), 
                                                ('border-bottom', '1px solid black'),
                                                ('width', '100%')]},
              {'selector': 'td', 'props': 'text-align: left;'}
          ])
```

V tabelo smo umestili samo spremenljivke, ki statistično značilno vplivajo na pojavnost poškodbe. To lahko vidimo v stolpcu kjer so navedene p vrednosti testa hi kvadratov. Stolpec 'Cramer's V' pove kakšna je povezanost med spremenljivko in prisotnostjo poškodbe. Amplitudo povezanosti lahko preverimo s pomočjo spodnje tabele.
