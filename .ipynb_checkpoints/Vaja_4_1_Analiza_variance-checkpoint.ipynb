## Poročilo vaja 4.1
#### Analiza varianece: anketa nogometnih trenerjev

V anketi nogometnih trenerjev nas zanima odziv trenerjev glede na njihove igralske izkušnje pred trenersko kariero.
1. Organiziramo trenerje glede na njihove igralske izkušnje na reprezentačnem nivoju, državnem nivoju in ostale. (Q12)
2. Glede na gornje skupine nas zanimajo (morebitne) razlike v odzivih na:
   - pomembne dejavnike za trenerski poklic (Q22)
   - dejavnike za razvoj talentov (Q31 in Q 37)
3. Izvedite ustrezno opisno statistiko in jo opremite tudi z grafi.
4. Izvedite ustrezno analizo in komentirajte rezultate.
5. Poročajte metodologijo in komentirajte rezultate.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from scipy.stats import kruskal

pd.options.display.max_columns = None
```

```python
path = os.getcwd()
df = pd.read_excel(os.path.join(path, "data\\anketa114621-2024-03-04.xlsx"))
```

```python
# Extract questions
df_questions22 = df.filter(regex='^Q22[a-z]$').copy()
df_questions31 = df.filter(regex='^Q31[a-z]$').copy()
df_questions37 = df.filter(regex='^Q37[a-z]$').copy()

questions22 = df_questions22.iloc[0].tolist()
questions31 = df_questions31.iloc[0].tolist()
questions37 = df_questions37.iloc[0].tolist()

df.drop([0], inplace=True)
```

#### 1. Razdelimo trenerje glede na njihove igralske izkušnje

```python
# remap Q12
skupine = {
    1: 'reprezentančni nivo',
    2: 'državni nivo',
    3: 'ostalo',
    4: 'ostalo'}
```

```python
# extract columns that correspond to questions 12, 22, 31, 37 only
df_dejavniki_trener = df.filter(regex='^Q12$|^Q22[a-z]$').copy()
df_razvoj_talentov = df.filter(regex='^Q12$|^Q31[a-z]$|^Q37[a-z]$').copy()
# clean all non-answered rows
df_dejavniki_trener = df_dejavniki_trener[~df_dejavniki_trener.isin([-3, -2, -1]).any(axis=1)]
df_razvoj_talentov = df_razvoj_talentov[~df_razvoj_talentov.isin([-3, -2, -1]).any(axis=1)]
```

```python
# replace numerals with likert scale strings
df_dejavniki_trener['Q12'] = df_dejavniki_trener['Q12'].map(skupine);
df_razvoj_talentov['Q12'] = df_razvoj_talentov['Q12'].map(skupine);
```

#### 2. Glede na gornje skupine nas zanimajo (morebitne) razlike v odzivih.
#### 3. Opisna statistika in ustrezni graf

```python
def convert_to_numeric(val):
    if isinstance(val, str) and '-' in val:
        try:
            a, b = map(float, val.split('-'))
            return (a + b) / 2
        except:
            return np.nan
    try:
        return float(val)
    except:
        return np.nan

df_trenerji = df[df.Q12 > 0].loc[:, ['Q2', 'Q7', 'Q12', 'Q13']]
df_trenerji['Q12_sorted'] = df_trenerji.Q12.map(skupine)
for col in ['Q2', 'Q7', 'Q13']:
    df_trenerji[col] = df_trenerji[col].apply(convert_to_numeric)

df_long = df_trenerji.melt(id_vars='Q12_sorted', value_vars=['Q2', 'Q7', 'Q13'],
                  var_name='question', value_name='value')
summary = df_long.groupby(['question', 'Q12_sorted'])['value'] \
                 .agg(['mean', 'std']) \
                 .round(1) \
                 .unstack()

summary.index = summary.index.map({
    'Q2': 'aktiven trener (leta)',
    'Q7': 'čas v klubu (ur/teden)',
    'Q13': 'aktiven nogometaš (leta)'
})

table_one = summary.style \
    .set_caption('Tabela 1: Povprečja in standardni odkloni po skupinah trenerjev') \
    .format('{:.2f}')\
    .set_table_styles([
        {'selector': 'caption', 'props': [('font-size', '16px'),
                                          ('font-weight', 'bold'),
                                          ('border-bottom', '1px solid black'),
                                          ('width', '100%')]},
        {'selector': 'td', 'props': [('text-align', 'center')]}
    ])
display(table_one)
```

Iz tabele 1 statistike lahko vidimo, da je so nogometaši na državnem in reprezentančnem nivoju vsaj dve leti več bili aktivni nogometaši kot trenerji, ki so igrali na nižjem nivoju, kot vsaj državnem. Prav tako lahko vidimo, da so v povprečju najdlje aktivni trenerji tisti, ki so kot aktivni nogometaši igrali na državnem nivoju. Največ časa pa v klubu preživijo trenerji, ki so v svoji karieri igrali na reprezentančnem nivoju. Iz spodnjega grafa lahko razberemo tudi, da je največ trenerjev tistih, ki so igrali na državnem (328) najmanj pa tistih, ki so igrali na reprezentančnem nivoju (65).

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Q12', data=df_dejavniki_trener)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', 
                (p.get_x() + p.get_width() / 2, height), 
                ha='center', va='bottom', fontsize=10)
plt.title('Igralske izkušnje trenerjev')
plt.xlabel('Igralske izkušnje')
plt.ylabel('Število trenerjev')
plt.tight_layout()
plt.show()
```

#### 4. Analiza
**Glede na pomembne dejavnike za trenerski poklic.**

Ker imamo opravka z neodvisnimi kategoričnimi spremenljivkami, uporabimo Kruskal-Wallisov test za primerjavo odzivov na vprašanje Q22. Rezultati so predstavljeni v naslednji tabeli.

```python

q22_cols = ['Q22a', 'Q22b', 'Q22c', 'Q22d', 'Q22e', 'Q22f', 'Q22g', 'Q22h', 
            'Q22i', 'Q22j', 'Q22k', 'Q22l', 'Q22m', 'Q22n', 'Q22o', 'Q22p', 
            'Q22q', 'Q22r', 'Q22s', 'Q22t', 'Q22u']

results = []

for i, q in enumerate(q22_cols):
    groups = [df_dejavniki_trener[df_dejavniki_trener['Q12'] == group][q].dropna() for group in ['reprezentančni nivo', 'državni nivo', 'ostalo']]
    if all(len(group) > 0 for group in groups):
        stat, p = kruskal(*groups)
        results.append({
            '': f'{q}: {questions22[i]}',
            'Statistic': stat, 
            'p-value': p
        })

kruskal_results = pd.DataFrame(results)
display(
    kruskal_results.style
    .set_caption('Q22: Kako pomembne so z ozirom na Vašo predstavo o trenerskem poklicu naslednji dejavniki?')\
    .format({col: "{:.2f}" for col in kruskal_results.select_dtypes(include='number').columns})
          .set_table_styles([
              {'selector': 'caption', 'props': [('font-size', '16px'), 
                                                ('font-weight', 'bold'), 
                                                ('border-bottom', '1px solid black'),
                                                ('width', '100%')]},
              {'selector': 'td', 'props': 'text-align: left;'}
          ])\
          .hide())
```

**Interpretacija:** H0 pravi, da ne obstajajo razlike med odgovori na katera so odgovorili trenerji iz posameznih skupin. Iz izračunanih p vrednosti lahko H0 zavrnemo v primeru odgovorov na naslednja vprašanja ***Q22e, Q22g, Q22j, Q22l, Q22n, Q22u***. 

**Zanimajo nas tudi razlike med skupinami v odzivu na dejavnike za razvoj talentov.** S Kruskal-Wallisovim testom tokrat preverimo odgovore na vprašanji Q31 in Q37.

```python
q31_cols = ['Q31a', 'Q31b', 'Q31c', 'Q31d', 'Q31e', 'Q31f', 'Q31g', 
            'Q31h', 'Q31i', 'Q31j', 'Q31k', 'Q31l', 'Q31m']
results = []

for i, q in enumerate(q31_cols):
    groups = [df_razvoj_talentov[df_razvoj_talentov['Q12'] == group][q].dropna() for group in ['reprezentančni nivo', 'državni nivo', 'ostalo']]
    if all(len(group) > 0 for group in groups):
        stat, p = kruskal(*groups)
        
        results.append({
            '': f'{q}: {questions31[i]}',
            'Statistic': stat, 
            'p-value': p
        })

kruskal_results = pd.DataFrame(results)
display(
    kruskal_results.style
    .set_caption('Q31: Ocenite katere značilnosti so najbolj pomembne za uspešno kariero igralca/igralke')\
    .format({col: "{:.2f}" for col in kruskal_results.select_dtypes(include='number').columns})
          .set_table_styles([
              {'selector': 'caption', 'props': [('font-size', '16px'), 
                                                ('font-weight', 'bold'), 
                                                ('border-bottom', '1px solid black'),
                                                ('width', '100%')]},
              {'selector': 'td', 'props': 'text-align: left;'}
          ])\
          .hide())
```

```python
q37_cols = ['Q37a', 'Q37b', 'Q37c', 'Q37d', 'Q37e', 'Q37f', 'Q37g', 'Q37h', 
            'Q37i', 'Q37j', 'Q37k', 'Q37l', 'Q37m', 'Q37n', 'Q37o']
results = []

for i, q in enumerate(q37_cols):
    groups = [df_razvoj_talentov[df_razvoj_talentov['Q12'] == group][q].dropna() for group in ['reprezentančni nivo', 'državni nivo', 'ostalo']]
    if all(len(group) > 0 for group in groups):
        stat, p = kruskal(*groups)
        results.append({
            '': f'{q}: {questions37[i]}',
            'Statistic': stat, 
            'p-value': p
        })

kruskal_results = pd.DataFrame(results)
display(
    kruskal_results.style
    .set_caption('Q37: Ocenite dejavnike, ki so pomembni za razvoj talenta')\
    .format({col: "{:.2f}" for col in kruskal_results.select_dtypes(include='number').columns})
    .set_table_styles([
              {'selector': 'caption', 'props': [('font-size', '16px'), 
                                                ('font-weight', 'bold'), 
                                                ('border-bottom', '1px solid black'),
                                                ('width', '100%')]},
              {'selector': 'td', 'props': 'text-align: left;'}
          ])\
          .hide())
```

**Interpretacija:** Opazimo lahko da v odzivih na vprašanje Q31 ni statistično značilnih razlik, medtem, ko se skupine v odzivih na vprašanje Q37 razlikujejo v odzivih na podvprašanja ***Q37k, Q37m in Q37o***. Obstajajo statistično značilne razlike med skupinami na odzive glede preprečevanja poškodb in pretreniranosti, zmožnostjo spoprijemanja s pritiski, ter distanciranjem od pričakovanj vpliva managerjev.

#### 5. Metodologija

Preučevali smo razliko med nogometnimi trenerji, ki so aktivno igrali na reprezentančnem, državnem ali drugem nivoju. Gledali smo ali obstajajo razlike med odgovori na vprašanja o pomembnih dejavnikih za trenerski poklic in dejavnikih za razvoj talentov glede na to na katerem nivoju so trenerji igrali, ko so bili še aktivni tekmovalci. Odgovori na vprašanja so bili v obliki Likertove skale (1-5). Za statistično analizo smo uporabili Kruskal-Wallisov test. Meja za statistično značilnost je p<0,05. Statistično analizo smo izvajali v Pythonu z uporabo scipy knjižnice. 

Pri pregledu odgovorov glede pomembnosti dejavnikov za trenerski poklic so trenerji različno pomembnost dodelili finančni stabilnosti kluba, možnostim napredovanja, predhodnjim izkušnjam in dosežkom, visokemu dohodku, aktivnem ukvarjanju z nogometom in odnosim z mediji. V ostalih odgovorih ni prišlo do statistično značilnih sprememb.

Pri pregledu odgovorov glede dejavnikov za razvoj talentov so se pokazale statistično značilne razlike pri mnenju glede preprečevanja poškodb in pretreniranosti, zmožnostjo spoprijemanja s pritiski in distanciranja od pričakovanj in vpliva managerjev.
